```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#1. Initialization
np.random.seed(42)
n_samples = 1000
```

```
#2. Generate Age (Normal Distribution)
age = np.random.normal(loc=35, scale=10, size=n_samples)
age = np.clip(age, 18, 65).astype(int)
```

```
 #3. Generate Gender (Categorical Distribution)
gender = np.random.choice(
    ["Male", "Female"],
    size=n_samples,
    p=[0.5, 0.5]
)
```

```
#4. Generate Education Level (Discrete Distribution)
education = np.random.choice(
    ["High School", "Bachelor", "Master", "PhD"],
    size=n_samples,
    p=[0.3, 0.4, 0.2, 0.1]
)
```

```
#5. Generate Annual Income (Log-Normal Distribution)
income = np.random.lognormal(mean=10, sigma=0.5, size=n_samples)
income = income.astype(int)
```

```
#6. Generate Credit Score (Normal Distribution)
credit_score = np.random.normal(loc=650, scale=50, size=n_samples)
credit_score = np.clip(credit_score, 300, 850).astype(int)
```

```
#7. Purchase Probability (Dependent Variable)
purchase_prob = (
    0.3 * (income / income.max()) +
    0.4 * (credit_score / 850) +
    0.3 * (age / 65)
)
purchase_prob = np.clip(purchase_prob, 0, 1)
```

```
#8. Purchase Decision (Bernoulli Distribution)
purchase = np.random.binomial(1, purchase_prob)
```

```
#9. Create Synthetic Dataset
synthetic_data = pd.DataFrame({
    "Age": age,
    "Gender": gender,
    "Education": education,
    "Annual_Income": income,
    "Credit_Score": credit_score,
    "Purchased": purchase
})
```

```
# 11. Basic Visualizations
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(age, bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")

plt.subplot(1, 3, 2)
plt.hist(income, bins=20, color='lightgreen', edgecolor='black')
plt.title("Income Distribution")

plt.subplot(1, 3, 3)
plt.hist(credit_score, bins=20, color='salmon', edgecolor='black')
plt.title("Credit Score Distribution")

plt.tight_layout()
plt.show()
```
